```python
import numpy as np

n = 10
x = np.array([10, 12, 18, 22, 29, 33, 37, 41, 25, 19])
y = np.array([8, 11, 17, 23, 28, 32, 35, 39, 22, 16])

mean_x = np.mean(x)
mean_y = np.mean(y)

cov = np.sum((x - mean_x) * (y - mean_y)) / (n - 1)
var_x = np.var(x, ddof=1)
var_y = np.var(y, ddof=1)

corr = cov / np.sqrt(var_x * var_y)

print('Vector X:', x)
print('Vector Y:', y)
print('Correlation:', round(corr, 4))
```
