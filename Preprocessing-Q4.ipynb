```python
import numpy as np

m = 4
n = 5

mat = np.array([
    [1.2, 3.5, 5.7, 2.1, 4.4],
    [6.8, 2.4, 3.6, 8.7, 5.9],
    [4.3, 7.1, 8.2, 1.4, 3.7],
    [9.8, 2.1, 4.5, 6.2, 7.0]
])

mat_t = mat.T

cov_mat = np.cov(mat_t, rowvar=False)

corr_mat = np.corrcoef(mat_t, rowvar=False)

print('Matrix (4x5):\n', mat, '\n')
print('Covariance Matrix (5x5):\n', cov_mat, '\n')
print('Correlation Matrix (5x5):\n', corr_mat, '\n')
```
