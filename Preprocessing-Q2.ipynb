```python
import numpy as np

n = 10
vectorX = np.array([12, 5, 9, 18, 22, 35, 44, 28, 13, 17])
vectorY = np.array([14, 7, 11, 20, 25, 38, 42, 30, 16, 19])

meanX = np.mean(vectorX)
meanY = np.mean(vectorY)

covXY = np.sum((vectorX - meanX) * (vectorY - meanY)) / (n - 1)

print("Covariance:", covXY)
```
